### Lectura de archivos ASCII

```r
data = read.table("./data/prc.txt", header=TRUE)
data
typeof(data)
class(data)
```

```r
names(data)
```

```r
lons=data$Lon
lats=data$Lat
prc=data$prc
cat("tipo:lons/lats/prc: ",typeof(lons), typeof(lats), typeof(prc))
cat("\ntipo:lons/lats/prc:",class(lons), class(lats), class(prc))
```

```r
n_elem =length(lons)
n_elem

for(i in 1:n_elem){
    cat("\n",i,"\t", lons[i],"\t", lats[1],"\t", prc[i])
}
```

```r
x<- 1:n_elem
max_y <- max(prc)

# Crear la primera curva
plot(x, prc, type = "b", frame = FALSE, pch = 20, lwd=4,
     las=1,ylim=c(9,max_y+5), col= "red",main="Sibinacocha", 
     xlab = "Enero", ylab = "mm/dia", font.axis=4, cex.axis=1.5,
     cex.lab=2, cex.main=3, col.main= "black",
     cex.sub = 0.75, font.sub = 3, col.sub = "green",
     col.lab ="darkblue")

lines(x, 0.5*prc, type="o", pch=22, lty=2, col="white")

legend("topright", legend=c("prc"),
       col=c("red"), lty = 1:2, cex=1.5,
      title="Estación 0001", text.font=4, bg='white',
      box.lty=2, box.lwd=2, box.col="white")
```

```r
f = function(t){
    s1 <- cos(2*pi*t)
    e1 <- exp(-t)
    return(s1*e1)
}

t<- seq(0, 5, by=0.1 )

# Crear la primera curva
plot(t, f(t), type = "b", frame = FALSE, pch = 19, 
     col= "red", xlab = "x", ylab = "y")

# Agregar una segunda curva
lines(t, 0.42*f(t), pch = 18, col = "blue", type = "b", lty = 2)

# Insertar leyenda
legend("topright", legend=c("Line 1", "Line 2"),
       col=c("red", "blue"), lty = 1:2, cex=0.8)
```
